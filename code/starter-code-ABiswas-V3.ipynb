#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
#numList must be in the form of a pandas dataframe slice.
def stdDev(numList):
    import numpy as np
    import pandas as pd

    numList=numList.dropna().to_numpy()
    var=[]
    meanVal=np.mean(numList)
    [var.append((i-meanVal)**2) for i in numList]
    standardDeviation=(np.sum(np.array(var))/len(var))**0.5
    
    return standardDeviation
````

````python
stdDev(df_combined2018[['Math_SAT2018']])
````

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(list_of_columns_as_dataframes, list_of_titles, list_of_xlabels,color,alpha,figsize,fontsize,filename):
    nrows = int(np.ceil(len(list_of_columns_as_dataframes)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns_as_dataframes): # Gives us an index value to get into all our lists
        ax[i].hist(column,alpha=alpha,facecolor=color,ec='white') # feel free to add more settings
        if len(list_of_columns_as_dataframes)%2==1: #remove empty subplot
            ax[len(list_of_columns_as_dataframes)].set_visible(False)
    for j, title in enumerate(list_of_titles):
        ax[j].set_title(title)
    for k, label in enumerate(list_of_xlabels):
        ax[k].set_xlabel(f'{label}')
        # Set titles, labels, etc here for each subplot
    fig.tight_layout()
    plt.rcParams['figure.figsize']=figsize
    plt.rcParams['font.size'] = fontsize
    plt.savefig(filename)
````
